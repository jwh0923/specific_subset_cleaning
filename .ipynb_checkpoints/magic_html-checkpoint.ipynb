```python
import os
import uuid
import traceback
from datetime import datetime
from llm_web_kit.input.datajson import DataJson
from llm_web_kit.extractor.extractor_chain import ExtractSimpleFactory
from func_timeout import FunctionTimedOut, func_timeout

os.environ["LLM_WEB_KIT_CFG_PATH"] = "/share/jiangwenhao/.llm-web-kit.jsonc"

def extract_data(_iter):
    from loguru import logger
    extractor_chain = ExtractSimpleFactory.create('/share/jiangwenhao/notebooks/定向子集专项/subset-spliz.jsonc')
    timeout_seconds = 10
    # 为每个分区创建唯一的错误日志文件
    # partition_id = str(uuid.uuid4())
    # current_time = datetime.now().strftime("%Y%m%d_%H%M%S")
    # error_log_path = f"s3://xyz-llm-users/xyz-users/yujia/CC-MAIN-2024-33/output/v002/error_logs/{current_time}_{partition_id}.json"
    # s3_doc_writer = S3DocWriter(path=error_log_path)

    for row in _iter:
        d = row.asDict()
        try:
            input_data = DataJson(d)
            data_e: DataJson = func_timeout(timeout_seconds, extractor_chain.extract, args=(input_data,))
            
            #data_e: DataJson = extractor_chain.extract(input_data)
            yield Row(**data_e.to_dict())
        except FunctionTimedOut as e1:
            d['__error'] = {
                "error_type":"TIMEOUT",
                "error_message": "extract function timeout",
                "traceback":"TIMEOUT"
            }
            yield Row(**d)
        except Exception as e:
            # 记录更详细的错误信息
            error_info = {
                "error_type": type(e).__name__,
                "error_message": str(e),
                "traceback": traceback.format_exc(),
            }
            logger.error(error_info)
            # s3_doc_writer.write(error_info)
            d['__error'] = error_info
            yield Row(**d)

    # 确保在分区处理完成后刷新并关闭文件
    # try:
    #     s3_doc_writer.flush()
    # except Exception as e:
    #     print(f"Error flushing error logs: {str(e)}")
        
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit, udf
from pyspark.sql.types import StringType, StructType, StructField, MapType
import uuid
import json
from datetime import datetime

# 初始化 SparkSession
spark = SparkSession.builder.appName("TransformJSON").getOrCreate()
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, lit, udf, from_json, schema_of_json
from pyspark.sql.types import StringType, StructType, MapType
import uuid
import json

# 初始化 SparkSession
spark = SparkSession.builder.appName("TransformJSON").getOrCreate()




# 定义可扩展的元数据处理函数
def transform_json(json_data):
    try:
        parsed_json = json.loads(json_data)
        meta_info = {
            "input_datetime": parsed_json.get("posttime", ""),
            "author": parsed_json.get("author", ""),
            "likes": parsed_json.get("likes", 0)
        }
        result = {
            "track_id": str(uuid.uuid4()),
            "dataset_name": "net-ease",
            "data_source_category": "HTML",
            "html": parsed_json.get("content", ""),
            "url": parsed_json.get("url", ""),
            "file_bytes": len(json_data.encode("utf-8")),
            "meta_info": json_dumps(meta_info)
        }
        return json.dumps(result)
    except Exception as e:
        print(f"Error processing JSON: {e}")
        return None

# 注册 UDF
transform_udf = udf(transform_json, StringType())

# 应用 UDF 转换每一行的 JSON 数据
result_df = sample_df.withColumn("transformed_json", transform_udf(col("value")))

# 展开转换后的 JSON 字符串为独立的列
schema = spark.read.json(result_df.rdd.map(lambda x: x.transformed_json)).schema
expanded_df = result_df.withColumn("transformed_json", from_json(col("transformed_json"), schema))
final_df = expanded_df.select("transformed_json.*")

# 查看结果
final_df.show(truncate=False)

# 保存为 JSON 文件
# final_df.write.json("/share/jiangwenhao/output.json", mode="overwrite")

```

```python
import zlib
import base64

def compress_and_decompress_str(input_data: Union[str, bytes], compress: bool = True, base: bool = False) -> Union[str, bytes]:
    compressed_bytes = base64.b64decode(input_data)
    decompressed_bytes = zlib.decompress(compressed_bytes)
    return decompressed_bytes.decode('utf-8')


def uncompress_html(_iter):
    for row in _iter:
        print(row)
        d = json_loads(row.value)
        raw_html = d.get("content", "")        
        d.pop("content", None)
        d["html"] = compress_and_decompress_str(raw_html)
        d["dataset_name"] = "netease"
        d["data_source_category"] = "html"                

        yield Row(value=json_dumps(d))


# sample_map_rdd = sample_df.rdd.mapPartitions(uncompress_html)
```

```python
acc = 

import os
os.environ["LLM_WEB_KIT_CFG_PATH"] = "/share/jiangwenhao/.llm-web-kit.jsonc"
import sys
# sys.path.append('/share/jiangwenhao/llm-webkit-mirror')
import uuid
import traceback
from datetime import datetime
from llm_web_kit.input.datajson import DataJson
from llm_web_kit.pipeline.pipeline_suit import PipelineSuit


def extract_data(_iter):
    pipesuit = PipelineSuit('/share/jiangwenhao/html_pipeline_formatter_disable.jsonc')
    # 为每个分区创建唯一的错误日志文件
    partition_id = str(uuid.uuid4())
    current_time = datetime.now().strftime("%Y%m%d_%H%M%S")
    error_log_path = f"s3://xyz-llm-users/xyz-users/yujia/CC-MAIN-2024-33/output/v002/error_logs/{current_time}_{partition_id}.json"
    s3_doc_writer = S3DocWriter(path=error_log_path)

    for row in _iter:
        try:
            d = json_loads(row.value)
            input_data = DataJson(d)
            data: DataJson = pipesuit.format(input_data)
            
            data_e: DataJson = pipesuit.extract(input_data)
            
            yield Row(value=json_dumps(data_e.to_json()))
        except Exception as e:
            # 记录更详细的错误信息
            error_info = {
                "error_type": type(e).__name__,
                "error_message": str(e),
                "traceback": traceback.format_exc(),
                "input_data": row.value if hasattr(row, 'value') else str(row),
                "timestamp": datetime.now().isoformat()
            }
            s3_doc_writer.write(error_info)

    # 确保在分区处理完成后刷新并关闭文件
    try:
        s3_doc_writer.flush()
    except Exception as e:
        print(f"Error flushing error logs: {str(e)}")

cc_main3_rdd = sample_df.rdd.mapPartitions(extract_data)

# cc_main_33_rdd.count()
```
